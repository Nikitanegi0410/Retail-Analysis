```python
df.drop_duplicates(subset=None,keep='first',inplace=True)
df.duplicated().sum()
```

```python
df['Postal Code'] = df['Postal Code'].astype('object')
```

```python
df = df.drop(['Postal Code'],axis = 1)  # drop postal code column
```

```python
df.columns
```

```python
df['Ship Mode'].value_counts() #value count of ship mode
```

```python
df['Ship Mode'].value_counts().plot.pie()
```

**Mostly preferrd standard class as compared to others.**

```python
df['Segment'].value_counts() # valuecount of segment
```

```python
plt.figure(figsize=(7,3))
sns.countplot(df['Segment'],palette='husl')
plt.show()
```

**Consumer segment is higher than another two segment.**

```python
df.groupby('Segment')[['Profit','Sales']].sum().plot.bar(color=['Maroon','black'],figsize=(8,5))
plt.ylabel('Profit/Loss and sales')
plt.show()
```

**Profit and sales are maximum in consumer segment and minimum in Home Office segment.**

```python
df['State'].value_counts()
```

```python
plt.figure(figsize=(10,8))
df['State'].value_counts().plot.pie(autopct = '%1.1f%%')
plt.show()
```

**Highest number of buyers are from California and New York and only one buyer is from Wyoming State.**

```python
ps = df.groupby('State')[['Sales','Profit']].sum().sort_values(by='Sales',ascending=False)
ps[:].plot.bar(color=['Green','Red'],figsize=(15,8))
plt.title('Profit/loss & Sales across states')
plt.xlabel('States')
plt.ylabel('Profit/loss & Sales')
plt.show()
```



*   Profit level is higher of California and New York as compared to other state.
*   Compared to other state texas, pennsylvania, Ohio is the state with rhe highest loss. 



```python
df['Category'].value_counts()  #valuecount of category
```

```python
plt.figure(figsize=(10,8))
df['Category'].value_counts().plot.pie(autopct = '%1.1f%%')
plt.show()
```

Office supplies is higher than other category.

```python
df.groupby('Category')[['Profit','Sales']].sum().plot.bar(color=['Red','Yellow'],alpha=0.9,figsize=(8,5))
plt.ylabel('Profit/Loss and sales')
plt.show()
```

Furniture is weak area to make profit.

```python
df['Sub-Category'].value_counts()
```

```python
plt.figure(figsize=(15,7))
sns.countplot(df['Sub-Category'],palette='husl')
plt.show()
```

```python
ps = df.groupby('Sub-Category')[['Sales','Profit']].sum().sort_values(by='Sales',ascending=False)
ps[:].plot.bar(color=['red','Blue'],figsize=(15,8))
plt.title('Profit/loss & Sales across states')
plt.xlabel('Sub-Category')
plt.ylabel('Profit/loss & Sales')
plt.show()
```




* High sale sub-category is phone and chairs but chair profit is less than phones.
* Huge loss in bookmark and tables.
* High profit in copies category.

```python
plt.figure(figsize=(10,8))
df['Region'].value_counts().plot.pie(autopct = '%1.1f%%')
plt.show()
```

Mostly buyers are from west region.

```python
plt.figure(figsize=(12,8))
plt.title('Segment wise Sales in each Region')
sns.barplot(x='Region',y='Sales',data=df,hue='Segment',order=df['Region'].value_counts().index,palette='rocket')
plt.xlabel('Region',fontsize=15)
plt.show()
```

Segment wise sales are almost same in every region.
In south region home office sale is higher than other two segment.

```python
df.groupby('Region')[['Profit','Sales']].sum().plot.bar(color=['Orange','Green'],figsize=(8,5))
plt.ylabel('Profit/Loss and sales')
plt.show()
```

**Profit and sales are maximum in west region and minimum in south region.**
